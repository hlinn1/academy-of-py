# PyCity Schools Analysis

* Percentage of students passing reading is generally higher than percentage of students passing math across the district. 

* Schools with lower budget allocated per student performed better than schools with higher per student budgets. Schools with highest per student budgets ($645-$675) performed worst compared to other schools. 

* Small and medium sized schools outperformed large schools. 93.55% of students passed Math in small schools, compared to 69.96% in large schools.
---

```python
# Dependencies and Setup
import pandas as pd
import numpy as np
pd.options.display.float_format = '{:.2f}'.format
# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head(10)
```

## District Summary

```python
total_schools = school_data_complete["school_name"].nunique()
total_students = school_data_complete["student_name"].count()
```

```python
grouped_school_data = school_data_complete.groupby(["school_name"])
school_budget = pd.DataFrame(grouped_school_data["budget"].mean())
#format it to show $ and commas
total_budget = school_budget["budget"].sum()
```

```python
average_math_score = school_data_complete["math_score"].mean()
average_reading_score = school_data_complete["reading_score"].mean()
overall_average_score = (average_math_score+average_reading_score)/2
```

```python
passing_math = school_data_complete.loc[(school_data_complete["math_score"] >= 70)]
percentage_passing_math = (len(passing_math) / total_students) * 100

passing_reading = school_data_complete.loc[(school_data_complete["reading_score"] >= 70)]
percentage_passing_reading = (len(passing_reading) / total_students) * 100

#overall passing rate should be the % of students who passed both? 
passing_both = len(school_data_complete[(school_data_complete.math_score >= 70) & (school_data_complete.reading_score >= 70)])
passing_both
overall_passing_rate = (passing_both / total_students) * 100
```

```python
district_dict = {"Total Schools": total_schools, "Total Students": total_students,
                "Total Budget": total_budget, "Average Math Score": average_math_score,
                "Average Reading Score": average_reading_score, "% Passing Math": percentage_passing_math,
                "% Passing Reading": percentage_passing_reading, "Overall Passing Rate": overall_passing_rate}
district_summary_df = pd.DataFrame(district_dict, index = [0])
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,}".format)
district_summary_df
```

## School Summary

```python
school_group = school_data_complete.groupby("school_name")
school_type = school_group["type"].unique()
school_size = school_group["size"].mean()
school_budget = school_group["budget"].mean()
per_student_budget = school_budget / school_size

school_ave_math_score = (school_group["math_score"].sum()) / school_size
school_ave_reading_score = (school_group["reading_score"].sum()) / school_size

mpass_per_school =  school_data_complete[school_data_complete["math_score"] >= 70].groupby("school_name").count()["student_name"]
rpass_per_school =  school_data_complete[school_data_complete["reading_score"] >= 70].groupby("school_name").count()["student_name"]

school_percent_pass_math = (mpass_per_school / school_size) * 100
school_percent_pass_reading = (rpass_per_school / school_size) * 100 
school_overall_passing_rate = (school_percent_pass_math + school_percent_pass_reading) / 2

```

```python
#create a dictionary of all the data above for the dataframe 
school_summary_dict = {"School Type": school_type, "Total Students": school_size,
                      "Total School Budget": school_budget, "Per Student Budget": per_student_budget,
                      "Average Math Score": school_ave_math_score, "Average Reading Score": school_ave_reading_score,
                      "% Passing Math": school_percent_pass_math, "% Passing Reading": school_percent_pass_reading,
                      "Overall Passing Rate": school_overall_passing_rate}

#create a data frame with the dictionary
school_summary_df = pd.DataFrame(school_summary_dict)

#clean up formatting
school_summary_df["Total School Budget"] = school_summary_df["Total School Budget"].map("${:,}".format)
school_summary_df["Per Student Budget"] = school_summary_df["Per Student Budget"].astype(float)
school_summary_df["Per Student Budget"] = school_summary_df["Per Student Budget"].map("{:.2f}".format)
school_summary_df["% Passing Math"] = school_summary_df["% Passing Math"].map("{:.2f}".format).astype(float)
school_summary_df["% Passing Reading"] = school_summary_df["% Passing Reading"].map("{:.2f}".format).astype(float)
school_summary_df["Overall Passing Rate"] = school_summary_df["Overall Passing Rate"].map("{:.2f}".format).astype(float)
school_summary_df.head()
```

## Top Performing Schools (By Passing Rate)

```python
top_schools = school_summary_df.sort_values("Overall Passing Rate", ascending = False)
top_schools.head()
```

## Bottom Performing Schools (By Passing Rate)

```python
bottom_schools = school_summary_df.sort_values("Overall Passing Rate", ascending = True)
bottom_schools.head()
```

## Math Scores by Grade

```python
ninth_grade_math = school_data_complete[school_data_complete["grade"] == "9th"].groupby("school_name")
ninth_grade_math = ninth_grade_math.mean()["math_score"].astype(float)

tenth_grade_math = school_data_complete[school_data_complete["grade"] == "10th"].groupby("school_name").mean()["math_score"]
eleventh_grade_math = school_data_complete[school_data_complete["grade"] == "11th"].groupby("school_name").mean()["math_score"]
twelfth_grade_math = school_data_complete[school_data_complete["grade"] == "12th"].groupby("school_name").mean()["math_score"]

math_by_grade_dict = {"9th": ninth_grade_math, "10th": tenth_grade_math,
                     "11th": eleventh_grade_math, "12th": twelfth_grade_math}
math_by_grade_df = pd.DataFrame(math_by_grade_dict)
math_by_grade_df
```

## Reading Score by Grade 

```python
ninth_grade_reading = school_data_complete[school_data_complete["grade"] == "9th"].groupby("school_name").mean()["reading_score"]
tenth_grade_reading = school_data_complete[school_data_complete["grade"] == "10th"].groupby("school_name").mean()["reading_score"]
eleventh_grade_reading = school_data_complete[school_data_complete["grade"] == "11th"].groupby("school_name").mean()["reading_score"]
twelfth_grade_reading = school_data_complete[school_data_complete["grade"] == "12th"].groupby("school_name").mean()["reading_score"]

reading_by_grade_dict = {"9th": ninth_grade_reading, "10th": tenth_grade_reading,
                     "11th": eleventh_grade_reading, "12th": twelfth_grade_reading}
reading_by_grade_df = pd.DataFrame(reading_by_grade_dict)
reading_by_grade_df
```

## Scores by School Spending

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
school_summary_copy = school_summary_df.copy()
school_summary_copy["Per Student Budget"] = school_summary_copy["Per Student Budget"].astype(float)
school_summary_copy["Spending Ranges (Per Student)"] = pd.cut(school_summary_copy["Per Student Budget"], spending_bins, labels=group_names)

mscore_by_spending = school_summary_copy.groupby("Spending Ranges (Per Student)").mean()["Average Math Score"]
rscore_by_spending = school_summary_copy.groupby("Spending Ranges (Per Student)").mean()["Average Reading Score"]
mpass_by_spending = school_summary_copy.groupby("Spending Ranges (Per Student)").mean()["% Passing Math"]
rpass_by_spending = school_summary_copy.groupby("Spending Ranges (Per Student)").mean()["% Passing Reading"]
opass_by_spending = school_summary_copy.groupby("Spending Ranges (Per Student)").mean()["Overall Passing Rate"]

scores_by_spending = {"Average Math Score": mscore_by_spending, "Average Reading Score": rscore_by_spending,
                     "% Passing Math": mpass_by_spending, "% Passing Reading": rpass_by_spending,
                     "Overall Passing Rate": opass_by_spending}

scores_by_spending_df = pd.DataFrame(scores_by_spending)
scores_by_spending_df
```

## Scores by School Size

```python
# Sample bins. Feel free to create your own bins.
size_bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
school_summary_size = school_summary_df.copy()

school_summary_size["School Size"] = pd.cut(school_summary_size["Total Students"], size_bins, labels=group_names)

mscore_by_size = school_summary_size.groupby("School Size").mean()["Average Math Score"]
rscore_by_size = school_summary_size.groupby("School Size").mean()["Average Reading Score"]
mpass_by_size = school_summary_size.groupby("School Size").mean()["% Passing Math"]
rpass_by_size = school_summary_size.groupby("School Size").mean()["% Passing Reading"]
opass_by_size = school_summary_size.groupby("School Size").mean()["Overall Passing Rate"]

scores_by_size = {"Average Math Score": mscore_by_size, "Average Reading Score": rscore_by_size, 
                 "% Passing Math": mpass_by_size, "% Passing Reading": rpass_by_size, 
                  "Overall Passing Rate": opass_by_size}
scores_by_size_df = pd.DataFrame(scores_by_size)
scores_by_size_df
```

## Scores by School Type

```python
school_summary_type = school_summary_df.copy()
school_summary_type["School Type"] = school_summary_type["School Type"].astype(str)

mscore_by_type = school_summary_type.groupby(["School Type"]).mean()["Average Math Score"]
rscore_by_type = school_summary_type.groupby(["School Type"]).mean()["Average Reading Score"]
mpass_by_type = school_summary_type.groupby(["School Type"]).mean()["% Passing Math"]
rpass_by_type = school_summary_type.groupby(["School Type"]).mean()["% Passing Reading"]
opass_by_type = school_summary_type.groupby(["School Type"]).mean()["Overall Passing Rate"]

scores_by_type = pd.DataFrame({"Average Math Score": mscore_by_type, "Average Reading Score": rscore_by_type,
                             "% Passing Math": mpass_by_type, "% Passing Reading": rpass_by_type,
                             "Overall Passing Rate": opass_by_type})
scores_by_type
```
